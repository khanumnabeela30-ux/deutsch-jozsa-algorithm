```python
%matplotlib inline

from qiskit import QuantumCircuit
from qiskit_aer import AerSimulator
from qiskit.visualization import plot_histogram
import matplotlib.pyplot as plt

# REAL Deutsch–Jozsa circuit that produces both 0 and 1 outcomes
qc = QuantumCircuit(2, 1)

# Step 1: Hadamard on both qubits
qc.h([0, 1])

# Step 2: Balanced oracle (CNOT)
qc.cx(1, 0)   # IMPORTANT: REVERSED ORDER TO MATCH OLD QISKIT 0.43 BEHAVIOR

# Step 3: Final Hadamard
qc.h(0)

# Step 4: Measurement
qc.measure(0, 0)

# Display circuit
qc.draw("mpl")
plt.show()

# Run with AerSimulator
sim = AerSimulator()
result = sim.run(qc, shots=1024).result()

counts = result.get_counts()
print("Deutsch–Jozsa Output:", counts)

# Show histogram
plot_histogram(counts)
plt.show()
```
